```python
import os

import requests

from bs4 import BeautifulSoup, element

project_path = os.getcwd()
img_path = os.path.join(project_path, "images")
```

```python
def fetch_page_content(url):
    """Fetches the HTML content of the page."""
    response = requests.get(url)
    response.raise_for_status()  # Raises an error for bad responses (e.g., 404, 500)
    return response.text
```

```python
url = 'https://ageofempires.fandom.com/wiki/War_Wagon_(Age_of_Empires_III)'
html_content = fetch_page_content(url)
```

```python
soup = BeautifulSoup(html_content, "html.parser")
infobox = soup.find('aside', class_='portable-infobox')
print(infobox.prettify())
```

```python
def download_unit_icon(infobox: element.Tag):
    """Downloads the icon of the unit."""
    # Find the icon image element in the infobox element
    icon = infobox.find('img', class_='pi-image-thumbnail', alt='Definitive')
    if icon and icon.has_attr('src'):
        # Fetch the icon image URL
        icon_url = icon['src']
        try:
            # Download the icon image
            icon_data = requests.get(icon_url).content
            # Get the icon name from the 'data-image-name' attribute
            icon_name = icon['data-image-name'].replace(" ", "_").lower()
            # Save the icon image to the images folder
            with open(os.path.join(img_path, icon_name), 'wb') as file:
                file.write(icon_data)
            print(f"Icono guardado como: {icon_name}")
        except Exception as e:
            print(f"Error al descargar la imagen: {e}")
```

```python
download_unit_icon(infobox)
```
