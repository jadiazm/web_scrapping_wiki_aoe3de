```python
%load_ext autoreload
%autoreload 2

import os
import sys
from collections import Counter
import requests
import json
from bs4 import BeautifulSoup as bs

project_path = os.path.abspath("..")

sys.path.append(project_path)
```

```python
def fetch_page_content(url):
    """Fetches the HTML content of the page."""
    response = requests.get(url)
    response.raise_for_status()
    # Raises an error for bad responses (e.g., 404, 500)
    return response.text
```

```python
aoe_wiki_url = "https://ageofempires.fandom.com/"
units_url = aoe_wiki_url + "wiki/Unit_(Age_of_Empires_III)"
html_content = fetch_page_content(units_url)
soup = bs(html_content, "html.parser")
main_page = soup.find('main', class_='page__main')
```

```python
content = soup.find('div', class_='mw-parser-output')
h2s = content.find_all('h2', recursive=False)
sections = [h2.find('span', class_='mw-headline', recursive=False).text for h2 in h2s]
```

```python
def get_section_units(section):
    for h2 in h2s:
        if h2.find('span', class_='mw-headline', recursive=False).text == section:
            ul = h2.find_next_sibling('ul')
            lis = ul.find_all('li', recursive=False)
            section_units = dict()
            for li in lis:
                anchor = li.find('a', recursive=False)
                if anchor is not None:
                    unit_name = anchor.text
                    unit_url = anchor.get('href')
                    section_units[unit_name] = aoe_wiki_url + unit_url
            return section_units
```

```python
all_hrefs = {section: get_section_units(section) for section in sections}
```
